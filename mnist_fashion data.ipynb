## LOADING DATA

```python
mnist = tf.keras.datasets.fashion_mnist
```

```python
(training_images , training_labels) , (test_images , test_labels) = mnist.load_data()
```

```python
training_images.shape
```

## NORMALIZING DATA

```python
training_images = training_images / 255.0
test_images = test_images / 255.0
```

## CREATING AND COMPILING MODEL

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(550 , activation = tf.nn.relu , input_shape = (28*28,)),
    tf.keras.layers.Dense(15, activation = tf.nn.softmax)
])
```

```python
model.compile(
    optimizer = tf.keras.optimizers.Adam(),loss = 'sparse_categorical_crossentropy', metrics = (['accuracy'])
)
```

## TRAINING

```python
model.fit(training_images,training_labels , epochs = 7 , batch_size = 128)
```

```python
model.evaluate(test_images,test_labels)
```
